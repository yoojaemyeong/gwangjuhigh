# 생성자와 소멸자 그리고 연산자 오버로딩

```python
# 생성자 
# __init__() 객체가 생성될 때 자동으로 호출
class Rabbit :
    shape = "" # 토끼 모양
    xPos = 0   # X위치
    yPos = 0   # Y위치
    
    def __init__(self,value='토끼'):
        self.shape = value

    def goto(self, x, y) :
        self.xPos = x
        self.yPos = y

## 객체 생성
rabbit1 = Rabbit()
print('rabbit1의 모양', rabbit1.shape)

rabbit2 = Rabbit('circle')
print('rabbit2의 모양', rabbit2.shape)

rabbit3 = Rabbit('triangle')
print('rabbit3의 모양', rabbit3.shape)

```

```python
# 소멸자
# __del__() 객체가 제거될 때 자동으로 호출

class Rabbit :
    shape = "" # 토끼 모양
    xPos = 0   # X위치
    yPos = 0   # Y위치
    
    def __init__(self,value='토끼'):
        self.shape = value
    
    def __del__(self):
        print(self.shape, " bye ~~")

    def goto(self, x, y) :
        self.xPos = x
        self.yPos = y

## 객체 생성 및 소멸
rabbit1 = Rabbit()
print('rabbit1의 모양', rabbit1.shape)
del(rabbit1)

rabbit2 = Rabbit('circle')
print('rabbit2의 모양', rabbit2.shape)
del(rabbit2)

rabbit3 = Rabbit('triangle')
print('rabbit3의 모양', rabbit3.shape)
del(rabbit3)
```

# 연산자 오버로딩

* \_\_add\_\_(self,other)  
* 그 외의 연산자도 재정의하면 됨

  

 

```python
# 연산자 오버로딩(1)

class Rabbit :
    shape = "" # 토끼 모양
    xPos = 0   # X위치
    yPos = 0   # Y위치
    
    def __init__(self,value='토끼'):
        self.shape = value
    
    def __del__(self):
        print(self.shape, " bye ~~")
        
    def __add__(self,other):
        print(self.shape,'와 ',other.shape,'는 친구가 되었습니다.')

    def goto(self, x, y) :
        self.xPos = x
        self.yPos = y

## 연산자 오버로딩
rabbit1 = Rabbit('산토끼')
print('rabbit1의 모양', rabbit1.shape)

rabbit2 = Rabbit('집토끼')
print('rabbit2의 모양', rabbit2.shape)

rabbit1 + rabbit2
```

```python
# 연산자 오버로딩(2)
# 선의 길이를 비교해서 선택하기
## 클래스 선언
class Line :
    length = 0

    def __init__(self, length=1) :  # 생성자
        self.length = length
    def __del__(self) :             # 소멸자
        print(self.length, '길이의 선 제거') 
    def __add__(self, other) :
        return self.length + other.length
    def __lt__(self, other) :
        return self.length < other.length
    def __eq__(self, other) :
        return self.length == other.length

## 메인 코드 부분, 선의 길이를 비교해 짧은 선 삭제하기
line1 = Line(10)
line2 = Line(5)

print('두 선의 합 : ', line1 + line2)

if line1 < line2 :
    print('선', line2.length, '가 더 깁니다.')
    del(line1)
elif line1 == line2 :
    print('두 선의 길이가 같습니다.')
else :
    print('선', line1.length,'이 더 깁니다.')
    del(line2)
```

# 클래스의 상속

* 기존 클래스의 속성과 행동(메소드)을 그대로 물려받는 새로운 클래스 생성  

> 새로운 클래스는 상속받은 속성과 행동 외에  
> 추가로 속성 또는 행동을 생성할 수 있다.  

  

```python
# 상속
class Rabbit :
    shape = ""
    xPos = 0
    yPos = 0
    def __init__(self,name='토끼'):
        self.shape=name
        
    def goto(self, x, y) :
        self.xPos = x
        self.yPos = y

class HouseRabbit(Rabbit) :
    owner = ""
    def eatFeed(self) :
        print(self.shape,"가 모이를 먹습니다",sep='')

class MountainRabbit(Rabbit) :
    mountain = ""
    def eatWildglass(self) :
        print(self.shape,"가 들풀을 먹습니다",sep='')

# 상속받은 객체 사용
hRabbit = HouseRabbit()
hRabbit.eatFeed()

mRabbit = MountainRabbit('산토끼')
mRabbit.eatWildglass()
```

# turtle 클래스를 상속받은 바다거북과 모래거북 만들기

  

```python
import turtle
# 바다거북: swim() 메소드 추가
class SeaTurtle(turtle.Turtle):
    name = ''
    body = None
    def __init__(self):
        self.body = turtle.Turtle("turtle")
        self.body.color("blue")
        self.name=self.body.shape()
    def swim(self,x,y):
        self.body.goto(x,y)

# 모래거북: walk() 메소드 추가
class SandTurtle(turtle.Turtle):
    name = ''
    body = None
    def __init__(self):
        self.body = turtle.Turtle("triangle")
        self.body.color("green")
        self.name = self.body.shape()
    def walk(self,x,y):
        self.body.goto(x,y)

#전역변수 선언
seaT, sandT  = None, None
seaT  = SeaTurtle()
sandT  = SandTurtle()

seaT.swim(100,100)
seaT.body.write(seaT.name, font=("Arial",20))

sandT.walk(-100,50)
sandT.body.write(sandT.name, font=("Arial",20))
```

# 토끼와 거북이 

* 거북이(turtle) 클래스를 상속받은 토끼 이동경로 체크하기  

  

```python
# 상속 구현(3)
import turtle
import random
## 클래스 선언부
class Rabbit :
    myTurtle = None
    def __init__(self, myTut) :
        self.myTurtle = myTut
        print("** 토끼와 거북이 **")
    def print_position(self) :
        print("토끼와 거북이 위치는 현재", self.myTurtle.xcor( ), ",", self.myTurtle.ycor(),"입니다.")
        
## 전역변수 선언부
myTut, myRab = None, None
colorList = ["red", "green", "blue", "black", "magenta", "orange", "gray"]
## 메인 코드부
turtle.setup(550, 550)
turtle.screensize(500, 500)

myTut = turtle.Turtle("turtle")
myRab = Rabbit(myTut)
myTut.pensize(5)
for _ in range(20) :
    x = random.randint(-250, 250)
    y = random.randint(-250, 250)
    color = random.choice(colorList)
    myTut.pencolor(color)
    myTut.goto(x,y)
    myRab.print_position( )

turtle.exitonclick( )
```
