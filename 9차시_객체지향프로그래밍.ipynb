```python
# 
# __del__()     

class Rabbit :
    shape = "" #  
    xPos = 0   # X
    yPos = 0   # Y
    
    def __init__(self,value=''):
        self.shape = value
    
    def __del__(self):
        print(self.shape, " bye ~~")

    def goto(self, x, y) :
        self.xPos = x
        self.yPos = y

##    
rabbit1 = Rabbit()
print('rabbit1 ', rabbit1.shape)
del(rabbit1)

rabbit2 = Rabbit('circle')
print('rabbit2 ', rabbit2.shape)
del(rabbit2)

rabbit3 = Rabbit('triangle')
print('rabbit3 ', rabbit3.shape)
del(rabbit3)
```

```python
#  (1)

class Rabbit :
    shape = "" #  
    xPos = 0   # X
    yPos = 0   # Y
    
    def __init__(self,value=''):
        self.shape = value
    
    def __del__(self):
        print(self.shape, " bye ~~")
        
    def __add__(self,other):
        print(self.shape,' ',other.shape,'  .')

    def goto(self, x, y) :
        self.xPos = x
        self.yPos = y

##  
rabbit1 = Rabbit('')
print('rabbit1 ', rabbit1.shape)

rabbit2 = Rabbit('')
print('rabbit2 ', rabbit2.shape)

rabbit1 + rabbit2
```

```python
#  (2)
#    
##  
class Line :
    length = 0

    def __init__(self, length=1) :  # 
        self.length = length
    def __del__(self) :             # 
        print(self.length, '  ') 
    def __add__(self, other) :
        return self.length + other.length
    def __lt__(self, other) :
        return self.length < other.length
    def __eq__(self, other) :
        return self.length == other.length

##   ,      
line1 = Line(10)
line2 = Line(5)

print('   : ', line1 + line2)

if line1 < line2 :
    print('', line2.length, '  .')
    del(line1)
elif line1 == line2 :
    print('   .')
else :
    print('', line1.length,'  .')
    del(line2)
```

```python
import turtle
# : swim()  
class SeaTurtle(turtle.Turtle):
    name = ''
    body = None
    def __init__(self):
        self.body = turtle.Turtle("turtle")
        self.body.color("blue")
        self.name=self.body.shape()
    def swim(self,x,y):
        self.body.goto(x,y)

# : walk()  
class SandTurtle(turtle.Turtle):
    name = ''
    body = None
    def __init__(self):
        self.body = turtle.Turtle("triangle")
        self.body.color("green")
        self.name = self.body.shape()
    def walk(self,x,y):
        self.body.goto(x,y)

# 
seaT, sandT  = None, None
seaT  = SeaTurtle()
sandT  = SandTurtle()

seaT.swim(100,100)
seaT.body.write(seaT.name, font=("Arial",20))

sandT.walk(-100,50)
sandT.body.write(sandT.name, font=("Arial",20))
```

```python
#  (3)
import turtle
import random
##  
class Rabbit :
    myTurtle = None
    def __init__(self, myTut) :
        self.myTurtle = myTut
        print("**   **")
    def print_position(self) :
        print("   ", self.myTurtle.xcor( ), ",", self.myTurtle.ycor(),".")
        
##  
myTut, myRab = None, None
colorList = ["red", "green", "blue", "black", "magenta", "orange", "gray"]
##  
turtle.setup(550, 550)
turtle.screensize(500, 500)

myTut = turtle.Turtle("turtle")
myRab = Rabbit(myTut)
myTut.pensize(5)
for _ in range(20) :
    x = random.randint(-250, 250)
    y = random.randint(-250, 250)
    color = random.choice(colorList)
    myTut.pencolor(color)
    myTut.goto(x,y)
    myRab.print_position( )

turtle.exitonclick( )
```
