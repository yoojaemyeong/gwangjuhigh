# 문자열 함수 : find, join, upper, lower, strip, lstrip, rstrip, replace, split, index

```python
#문자열 함수 실습1
ex1 = 'Now is better than never'
print('th의 위치:',ex1.find('th'))
print('p 문자찾기:',ex1.find('p'))
print('e 문자찾기:',ex1.find('e',10))
#print('p 문자찾기:',ex1.index('p'))
print('th의 위치:',ex1.index('th'))
```

```python
#문자열 함수 실습2
ex2 = '  Abcd  '
print('대문자로:',ex2.upper())
print('소문자로:',ex2.lower())
print('왼쪽 빈공간 지우기:',ex2.lstrip())
print('오른쪽 빈공간 지우기:',ex2.rstrip())
print('오른쪽 빈공간 지우기:',ex2.strip())
```

```python
#문자열 함수 실습3
ex3 = 'Python is difficult'
print('difficult -> funny:',ex3.replace('difficult','funny'))
print('문자열을 리스트로:',ex3.split())
ex4 = "a,b,c,d"
print(ex4)
print(ex4.split(','))
```

## 문자는 ASCII 코드표에 의해 연산

### [ASCII 코드표](https://namu.wiki/w/%EC%95%84%EC%8A%A4%ED%82%A4%20%EC%BD%94%EB%93%9C)      
  

* ASCII 코드표에 의해 알파벳 문자를 숫자로 변환해주는 함수 : ord('문자')
  > ord('A')  
  > 결과: 65
* ASCII 코드표에 의해 숫자를 알파벳 문자로 변환해주는 함수 : chr(숫자)
  > chr(97)  
  > 결과: a  

```python
# ASCII :: 문자 -> 숫자
#대문자
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in range(0,len(alphabet)):
    if alphabet[i]=="Z":
        print("%c(%d)" % ( alphabet[i], ord(alphabet[i])))
    else:
        print("%c(%d)" % ( alphabet[i], ord(alphabet[i])), end=", ")
#소문자         
alphabet2 = alphabet.lower()        
for i in range(0,len(alphabet2)):
    if alphabet2[i]=="z":
        print("%c(%d)" % ( alphabet2[i], ord(alphabet2[i])))
    else:
        print("%c(%d)" % ( alphabet2[i], ord(alphabet2[i])), end=", ")

#숫자를 알파벳 문자로
for i in range(97,123):
    if i==122:
        print("%d(%c)" % (i,chr(i)))
    else:
        print("%d(%c)" % (i,chr(i)), end=", ")
        
```

# 영역 표시하기
<img src=attachment:b619bd79-ed8f-49ea-b228-4e59ade10668.png width=70% height=70% /> 


```python
#코드 작성
import turtle

turtle.shape("turtle")
turtle.penup()

while True :
    x = input("X위치:")
    if x == '':
        print('프로그램을 종료합니다.')
        break
    else:
        x = int(x)    
    y = int(input("Y위치:"))
    
    text= input("영역 표시 메시지 입력:")

    turtle.goto(x,y)
    turtle.write(text, font=("Arial", 30))

turtle.exitonclick()
```
