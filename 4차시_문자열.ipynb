#   : find, join, upper, lower, strip, lstrip, rstrip, replace, split, index

```python
#  1
ex1 = 'Now is better than never'
print('th :',ex1.find('th'))
print('p :',ex1.find('p'))
print('e :',ex1.find('e',10))
#print('p :',ex1.index('p'))
print('th :',ex1.index('th'))
```

```python
#  2
ex2 = '  Abcd  '
print(':',ex2.upper())
print(':',ex2.lower())
print('  :',ex2.lstrip())
print('  :',ex2.rstrip())
print('  :',ex2.strip())
```

```python
#  3
ex3 = 'Python is difficult'
print('difficult -> funny:',ex3.replace('difficult','funny'))
print(' :',ex3.split())
ex4 = "a,b,c,d"
print(ex4)
print(ex4.split(','))
```

```python
# 
import turtle

turtle.shape("turtle")
turtle.penup()

while True :
    x = input("X:")
    if x == '':
        print(' .')
        break
    else:
        x = int(x)    
    y = int(input("Y:"))
    
    text= input("   :")

    turtle.goto(x,y)
    turtle.write(text, font=("Arial", 30))

turtle.exitonclick()
```
