```python
#  1
ex1 = 'Now is better than never'
print('th :',ex1.find('th'))
print('p :',ex1.find('p'))
print('e :',ex1.find('e',10))
#print('p :',ex1.index('p'))
print('th :',ex1.index('th'))
```

```python
#  2
ex2 = '  Abcd  '
print(':',ex2.upper())
print(':',ex2.lower())
print('  :',ex2.lstrip())
print('  :',ex2.rstrip())
print('  :',ex2.strip())
```

```python
#  3
ex3 = 'Python is difficult'
print('difficult -> funny:',ex3.replace('difficult','funny'))
print(' :',ex3.split())
ex4 = "a,b,c,d"
print(ex4)
print(ex4.split(','))
```

##  ASCII   

### [ASCII ](https://namu.wiki/w/%EC%95%84%EC%8A%A4%ED%82%A4%20%EC%BD%94%EB%93%9C)      
  

* ASCII        : ord('')
  > ord('A')  
  > : 65
* ASCII        : chr()
  > chr(97)  
  > : a  

```python
# ASCII ::  -> 
#
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in range(0,len(alphabet)):
    if alphabet[i]=="Z":
        print("%c(%d)" % ( alphabet[i], ord(alphabet[i])))
    else:
        print("%c(%d)" % ( alphabet[i], ord(alphabet[i])), end=", ")
#         
alphabet2 = alphabet.lower()        
for i in range(0,len(alphabet2)):
    if alphabet2[i]=="z":
        print("%c(%d)" % ( alphabet2[i], ord(alphabet2[i])))
    else:
        print("%c(%d)" % ( alphabet2[i], ord(alphabet2[i])), end=", ")

#  
for i in range(97,123):
    if i==122:
        print("%d(%c)" % (i,chr(i)))
    else:
        print("%d(%c)" % (i,chr(i)), end=", ")
        
```

```python
# 
import turtle

turtle.shape("turtle")
turtle.penup()

while True :
    x = input("X:")
    if x == '':
        print(' .')
        break
    else:
        x = int(x)    
    y = int(input("Y:"))
    
    text= input("   :")

    turtle.goto(x,y)
    turtle.write(text, font=("Arial", 30))

turtle.exitonclick()
```
