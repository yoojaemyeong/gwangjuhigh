````python
# 버튼 컨트롤 + pack()
from tkinter import *
import tkinter.messagebox as msgbox

## 함수 선언부
def myFunc() :
    msgbox.showinfo("버튼1", "버튼 클릭")

## 메인 코드부
win = Tk()
win.geometry('300x100')

button1 = Button(win,text="클릭하세요",fg="red",bg="yellow", command=myFunc)
button1.pack()

win.mainloop()
````

````python
# 체크버트 컨트롤 + pack()
from tkinter import * 
import tkinter.messagebox as msgbox

## 함수 선언부
def myFunc() :
    if chk.get() == 0 :
        msgbox.showinfo("", "체크버튼 OFF 네요.")
    else :
        msgbox.showinfo("", "체크버튼 ON 이네요.")

## 메인 코드부
win = Tk()
win.geometry('300x100')

chk = IntVar()
cb1 = Checkbutton(win, text="클릭하세요", variable=chk, command=myFunc)
cb1.pack()

win.mainloop()
````

Tkinter에서 사용하는 변수 클래스: StringVar, IntVar, DoubleVar    
변수클래스는 GUI프로그램에서 사용자 입력을 관리 및 값 변경  
변수클래스 참조: https://docs.python.org/3/library/tkinter.html  


````python
# 라디오버튼 컨트롤 + pack()
from tkinter import *
win = Tk()
## 함수 선언부
def myFunc() :
    if myVar.get() == 1 :
        label2.configure(text="벤츠")
    elif myVar.get() == 2 :
        label2.configure(text="BMW")
    else :
        label2.configure(text="아우디")

## 메인 코드부
win = Tk()
win.geometry('300x200')
label1 = Label(win, text="자신이 선호하는 차량을 선택하세요", fg="blue").pack(anchor='center')  # 라벨의 위치 anchor : w,e,s,n,center..
myVar = IntVar(win,1)  #라디오버튼 기본 선택 정의
rb1=Radiobutton(win,text="벤츠",variable=myVar,value=1,command=myFunc).pack(anchor='w')
rb2=Radiobutton(win,text="BMW",variable=myVar,value=2,command=myFunc).pack(anchor='w')
rb3=Radiobutton(win,text="아우디",variable=myVar,value=3,command=myFunc).pack(anchor='w')
label2 = Label(win, text="선택한 차량 : ", fg="red")
label2.pack()

win.mainloop()
````

````python
# 마우스 왼쪽, 가운데, 오른쪽 버튼 클릭
from tkinter import *
from tkinter import messagebox as msgbox

# 함수 정의
def clickMouse1(event):
    msgbox.showinfo("마우스", "마우스 왼쪽 버튼을 클릭하였습니다.")
def clickMouse2(event):
    msgbox.showinfo("마우스", "마우스 가운데 버튼을 클릭하였습니다.")
def clickMouse3(event):
    msgbox.showinfo("마우스", "마우스 오른쪽 버튼을 클릭하였습니다.")

# 메인코드: Entry Point
if __name__ == '__main__':
    win = Tk()
    win.bind("<Button-1>", clickMouse1)
    win.bind("<Button-2>", clickMouse2)
    win.bind("<Button-3>", clickMouse3)
    win.mainloop()
    
````

## event 정보 확인(1)
- 마우스 클릭 위치(x,y)
- 마우스 버튼 번호(num)

  
  

````python
# event 정보 확인(2) : 마우스 버튼번호(num), 위치(x,y)
from tkinter import *
from tkinter import messagebox as msgbox

# 함수 정의
def clickMouse(event) :
    if event.num == 1:
    	msgbox.showinfo("event정보확인",f'왼쪽 버튼({event.num}) : 위치({event.x},{event.y})') 
    elif event.num == 2:
    	msgbox.showinfo("event정보확인",f'가운데 버튼({event.num}) : 위치({event.x},{event.y})')
    elif event.num == 3:
    	msgbox.showinfo("event정보확인",f'오른쪽 버튼({event.num}) : 위치({event.x},{event.y})')

# 메인코드 : Entry Point
if __name__ == '__main__':
    win = Tk()
    win.bind("<Button>", clickMouse)
    win.mainloop()
````

## 캔버스 Canvas
* 도형 그리기(선, 원, 사각형)
```python
# 관련 함수
  canvas.create_line(x1,y1,x2,y2,width=3,fill='blue')
  canvas.create_oval(x1,y1,x2,y2,outline='blue')
  canvas.create_rectangle(x1,y1,x2,y2,outline='green',fill='yellow')
  canvas.create_polygon([[x1,y1],[x2,y2],[x3,y3],[x4,y4],[x5,y5]],outline='blue',fill='green')
  canvas.create_text([10,10],text='test',font=('굴림',15))
```

  

````python
# Canvas 도형 그리기
from tkinter import *

# 메인코드 : Entry Point
if __name__ == '__main__':
    win = Tk()
    win.geometry("500x500")
    # 도형만들기
    canvas = Canvas(win,height=400,width=400)
    canvas.pack(side=BOTTOM,anchor=CENTER)
    canvas.create_line(0,0,50,50,width=3,fill='blue')
    canvas.create_text([25,25],text='line',font=('굴림',15))
    canvas.create_oval(50,50,150,150,outline='red')
    canvas.create_text([100,100],text='oval',font=('굴림',15))
    canvas.create_rectangle(151,151,400,400,width=5,outline='green',fill='yellow')
    canvas.create_text([250,350],text='rectangle',font=('굴림',15))
    canvas.create_polygon([[100,100],[100,200],[150,300],[200,300],[250,250],[250,200],[200,100]],outline='blue',fill='ivory')
    canvas.create_text([200,200],text='polygon',font=('굴림',15))
    # 실행
    win.mainloop()
````

## 마우스 클릭으로 도형 그리기
* 조건
  > Canvas 경계를 넘어서지 않기  
  > 마우스 왼쪽 버튼 클릭 : 지름이 50인 빨강 테두리 원  
  > 마우스 오른쪽 버튼 클릭: 지름이 70인 파랑 테두리 사각형  
  

````python
# 마우스 클릭으로 도형 그기기
from tkinter import *

# 함수 정의
def clickMouse(event) :        
    if event.num == 1:   # 마우스 왼쪽 버튼
        if event.x <= 450 and event.y <= 450:
            canvas.create_oval(event.x,event.y,event.x+50,event.y+50,outline='red')
        elif event.x > 450 and event.y <= 450:
            canvas.create_oval(event.x,event.y,500,event.y+50,outline='red')
        elif event.x <= 450 and event.y > 450:
            canvas.create_oval(event.x,event.y,event.x+50,500,outline='red')
        else:
            canvas.create_oval(event.x,event.y,500,500,outline='red')
    elif event.num == 2: # 마우스 가운데 버튼
    	txt=f'가운데 버튼({event.num}) : 위치({event.x},{event.y})'
    elif event.num == 3: # 마우스 오른쪽 버튼
        if event.x <= 430 and event.y <= 430:
            canvas.create_rectangle(event.x,event.y,event.x+70,event.y+70,outline='blue')
        elif event.x > 430 and event.y <= 430:
            canvas.create_rectangle(event.x,event.y,500,event.y+70,outline='blue')
        elif event.x <= 430 and event.y > 430:
            canvas.create_rectangle(event.x,event.y,event.x+70,500,outline='blue')
        else:
            canvas.create_rectangle(event.x,event.y,500,500,outline='blue')    

# 메인코드 : Entry Point
if __name__ == '__main__':    
    win = Tk()
    win.geometry("500x500")    
    canvas = Canvas(win, height=500,width=500)
    canvas.pack(anchor=CENTER)    
    win.bind("<Button>", clickMouse)
    win.mainloop()
````
