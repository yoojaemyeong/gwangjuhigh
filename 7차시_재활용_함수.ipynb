````python
#  
def myCalc(num1,num2,op):            
    if op=='+':
        return num1+num2
    elif op=='-':
        return num1-num2
    elif op=='*':
        return num1*num2
    elif op=='/':
        if num2==0:
            print('0  ')
            return 0
        else:
            return num1/num2
    else:
        print('  .') 
        return 0

print('=====  =====')
op = input('(+,-,*,/) : ')
num1 = int(input('  : '))
num2 = int(input('  : '))
print(' (',num1,op,num2,'):',myCalc(num1,num2,op))
````

````python
#   
def para_func(v1, v2, v3=0):
    return v1+v2+v3

print(' 2  :',para_func(5,10))
print(' 3  :',para_func(5,10,15))
````

````python
#     
def sum2(*args):
    x=0
    for i in args:
        x+=i
    return x

print('sum2   :', sum2(1,2,3,4,5,6,7,8,9))
````

````python
#       : global
def func1():
    global a
    a = 10
    print('func1() a :',a)

def func2():
    print('func2() a :',a)

#
a = 30

# 
if __name__ == '__main__':
    func1()
    func2()
````

````python
#    
# 1.   8  False 
# 2.  8      : True  False
def checkPassword(pwd):
    if len(pwd) < 8:
        return False
    if pwd.isalnum():
        return True
    else:
        return False

# 
password = input('  : ')

if checkPassword(password):
    print('  .')
else:
    print('   ')      
````

````python
#     
import turtle
import random

def getXYAS() :
    x, y, size = 0, 0, 0
    x = random.randint(-250, 250)
    y = random.randint(-250, 250)
    size = random.randint(10, 50)
    return x, y, size

##    ##
koreanStr = """        ,
          
    .       
          
 .
"""
colorList = ["red", "green", "blue", "black", "magenta", "orange", "gray", "pink", "yellow","lightgreen"]
tX, tY, txtSize = 0,0,0

##    ##
turtle.shape('turtle')
turtle.setup(550, 550)
turtle.screensize(500, 500)
turtle.penup( ) #   
turtle.speed(5)
for ch in koreanStr :
    tX, tY, txtSize = getXYAS( )
    color = random.choice(colorList)
    turtle.goto(tX,tY)
    turtle.pencolor(color)
    turtle.write(ch, font=('', txtSize, 'bold')) #  

turtle.exitonclick( )
````
