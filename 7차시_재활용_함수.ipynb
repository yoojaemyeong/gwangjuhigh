## 함수를 사용한 계산기 구현

1. 함수명: myCalc()
2. 계산을 위한 기호 +, -, *, /를 입력  
3. 두 수를 입력  
4. 결과 출력  

> [실행결과]
> 입력방식(+,-,*,/) 선택:  -    
> 첫 번째 숫자 입력:  55  
> 두 번째 숫자 입력:  77  
> 계산 결과( 55 - 77 ): -22  

````python
# 계산기 구현
def myCalc(num1,num2,op):            
    if op=='+':
        return num1+num2
    elif op=='-':
        return num1-num2
    elif op=='*':
        return num1*num2
    elif op=='/':
        if num2==0:
            print('0으로 나눈 에러')
            return 0
        else:
            return num1/num2
    else:
        print('입력한 연산자는 없습니다.') 
        return 0

print('===== 계산기 =====')
op = input('입력방식(+,-,*,/) 선택: ')
num1 = int(input('첫 번째 숫자입력: '))
num2 = int(input('두 번째 숫자입력: '))
print('계산 결과(',num1,op,num2,'):',myCalc(num1,num2,op))
````

````python
# 매개변수 기본값 설정하기
def para_func(v1, v2, v3=0):
    return v1+v2+v3

print('매개변수 2개 입력한 결과:',para_func(5,10))
print('매개변수 3개 입력한 결과:',para_func(5,10,15))
````

````python
# 인수 개수가 정해지지 않는 경우
def sum2(*args):
    x=0
    for i in args:
        x+=i
    return x

print('sum2 함수 실행 결과:', sum2(1,2,3,4,5,6,7,8,9))
````

## 전역변수 

* 프로그램 전체(global)에서 사용되는 변수  



````python
# 지역변수와 전역변수의 이름이 같은 경우
def func1():
    a = 10
    print('func1()에서 a의 값:',a)

def func2():
    print('func2()에서 a의 값:',a)

#전역변수
a = 30

#메인 코드
if __name__ == '__main__':
    func1()
    func2()
````

````python
# 함수에서 지역변수 대신 전역변수를 사용하고자 할때 : global
def func1():
    global a
    a = 10
    print('func1()에서 a의 값:',a)

def func2():
    print('func2()에서 a의 값:',a)

#전역변수
a = 30

#메인 코드
if __name__ == '__main__':
    func1()
    func2()
````

````python
# 비밀번호 규칙 확인 
# 1. 비밀번호 길이가 8글자 미만 False 
# 2. 비밀번호 8글자 이상이면서 문자 또는 숫자로 구성 : True 그렇지않으면 False
def checkPassword(pwd):
    if len(pwd) < 8:
        return False
    if pwd.isalnum():
        return True
    else:
        return False

# 전역변수
password = input('새로운 비밀번호를 입력하세요: ')

if checkPassword(password):
    print('비밀번호가 올바르게 생성되었습니다.')
else:
    print('비밀번호 규칙에 맞지 않습니다')      
````

# (실습) 훈민정음 그래픽 디자인 

* 훈민정음 글자를 랜덤한 위치(x,y), 글자크기(size), 글자색상(color)을 조합하여 구성하기   



````python
# 훈민정음 해례본 현대어 그래픽 디자인
import turtle
import random

def getXYAS() :
    x, y, size = 0, 0, 0
    x = random.randint(-250, 250)
    y = random.randint(-250, 250)
    size = random.randint(10, 50)
    return x, y, size

## 전역변수 선언 부분 ##
koreanStr = """ 나라의 말이 중국과 달라 문자와 서로 통하지 아니하므로,
이런 까닭으로 어리석은 백성이 이르고자 하는 바가 있어도 마침내 제 뜻을
능히 펴지 못하는 사람이 많으니라. 내가 이를 위하여 가엾게 여겨 새로 
스물여덟 자를 만드니 사람마다 하여금 쉽게 익혀 날로 쓰기 편하게 하고자
할 따름이니라.
"""
colorList = ["red", "green", "blue", "black", "magenta", "orange", "gray", "pink", "yellow","lightgreen"]
tX, tY, txtSize = 0,0,0

## 메인 코드 부분 ##
turtle.shape('turtle')
turtle.setup(550, 550)
turtle.screensize(500, 500)
turtle.penup( ) # 이동만을 위한 설정
turtle.speed(5)
for ch in koreanStr :
    tX, tY, txtSize = getXYAS( )
    color = random.choice(colorList)
    turtle.goto(tX,tY)
    turtle.pencolor(color)
    turtle.write(ch, font=('맑은고딕', txtSize, 'bold')) # 한글 쓰기

turtle.exitonclick( )
````
